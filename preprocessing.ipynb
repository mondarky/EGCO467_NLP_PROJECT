```python
import pandas as pd
import numpy as np
import matplotlib as plt
import json
import jsonlines
```

# Reading json file and data cleaning

```python
data = []
with jsonlines.open('Dataset/output_20201120.json') as f:
    for line in f:
        data.append(line)
```

```python
print(data[0])
print(type(data),type(data[0]))
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
for i, l in enumerate(df['label']):
 print(i,l,type(l))
```

```python
#https://towardsdatascience.com/dealing-with-list-values-in-pandas-dataframes-a177e534f173
def to_1D(series):
    return pd.Series([x for _list in series for x in _list])

def boolean_df(item_lists, unique_items):
# Create empty dict
    bool_dict = {}
    
    # Loop through all the tags
    for i, item in enumerate(unique_items):
        
        # Apply boolean mask
        bool_dict[item] = item_lists.apply(lambda x: item in x)
            
    # Return the results as a dataframe
    return pd.DataFrame(bool_dict)
```

```python
to_1D(df['label']).value_counts()
```

```python
unique_items = set()
for list in df['label']:
    for word in list:
        if word not in unique_items:
            unique_items.add(word)
print(unique_items)
```

```python
unique_items
label_bool = boolean_df(df['label'], unique_items)
print(label_bool)
```

```python
df['text'] = df['text'].str.replace('\n',"", regex=True)
```

# tokenize with deepcut

```python
import deepcut
```

```python
%%time
i = 0
for x in df['text']:
    df.iloc[i]['text'] = deepcut.tokenize(df.iloc[i]['text'])
    i += 1
```

```python
df
```

```python
result = pd.concat([df, label_bool], axis=1, sort=False)
result = result.drop(columns=['label'])
```

```python
result
```

```python
result.to_pickle('Dataset/tokenize.pkl')
```

# word2vec with fasttext

```python
import fasttext
```

```python
#wiki word vector
#https://fasttext.cc/docs/en/pretrained-vectors.html
ft = fasttext.load_model('wiki.th.bin')
```

```python
%%time
i = 0
for x in df['text']:
    df.iloc[i]['text'] = ft.get_sentence_vector(df.iloc[i]['text'])
    i += 1
```

```python
result2 = pd.concat([df, label_bool], axis=1, sort=False)
result2 = result2.drop(columns=['label'])
```
