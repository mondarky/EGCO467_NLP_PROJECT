```python
import pandas as pd
import numpy as np
import matplotlib as plt
import json
import jsonlines
```

# Reading json file and data cleaning

```python
data = []
with jsonlines.open('Dataset/output_20201120.json') as f:
    for line in f:
        data.append(line)
```

```python
index = []
count = 0
for line in data:
    if (line['label'] == []):
        index.append(count)
    count+=1
    
# reverse a list
index = index[::-1]

# delete empty label
for n in index:
    del data[n]
```

```python
df = pd.DataFrame(data)
df
```

```python
for i, l in enumerate(df['label']):
    print(i,l,type(l))
```

```python
#https://towardsdatascience.com/dealing-with-list-values-in-pandas-dataframes-a177e534f173
def to_1D(series):
    return pd.Series([x for _list in series for x in _list])

def boolean_df(item_lists, unique_items):
# Create empty dict
    bool_dict = {}
    
    # Loop through all the tags
    for i, item in enumerate(unique_items):
        
        # Apply boolean mask
        bool_dict[item] = item_lists.apply(lambda x: item in x)
            
    # Return the results as a dataframe
    return pd.DataFrame(bool_dict)
```

```python
to_1D(df['label']).value_counts()
```

```python
unique_items = set()
for list in df['label']:
    for word in list:
        if word not in unique_items:
            unique_items.add(word)
print(unique_items)
```

```python
unique_items
label_bool = boolean_df(df['label'], unique_items)
print(label_bool)
```

```python
blacklist = set('!#^<>=&@/()-,[]\"\'\n?+$*:_.%~')
for i in blacklist:
    df['text'] = df['text'].str.replace(i,"", regex=True)
```

```python
result = pd.concat([df, label_bool], axis=1, sort=False)
result = result.drop(columns=['label'])
```

# Tokenize text

```python
text = result['text'].tolist()
```

```python
import deepcut
with open('Dataset/tokenize.txt', 'w', encoding="utf-8") as f:
    for frame in text:
        joint_line = ''
        tokens = deepcut.tokenize(frame)
        for token in tokens:
          joint_line += (token + ' ')
        joint_line += ('\n')
        f.write(joint_line)
```
